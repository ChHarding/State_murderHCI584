```python
# read in US counties DB
df_counties = pd.read_csv("uscounties.csv")
df_counties.head(5)
```

```python
# find a given County (with state) in county DB (to get its lat/long)

state = "IL" # would later come from person DB
county = "Cook"

query_string = f'state_id == "{state}" and county == "{county}"'  # the " around the values are needed!

cnty = df_counties.query(query_string)
print(cnty)
lat, long = float(cnty["lat"]), float(cnty["lng"])
print(lat, long)
```
