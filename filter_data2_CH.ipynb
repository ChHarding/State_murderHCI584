```python
# filter/clean up data frame and write into filtered.csv
import pandas as pd
df = pd.read_csv('innocence_project_stats.csv')
col_names = list(df)
col_names
```

```python
# reorder and drop columns
df = df[['LastName', 'FirstName', 'Age', 'Race', 'Crime', 'Sentence', 'Convicted', 'Exonerated', 'DNA', "ST", "County"]]
print(list(df))
```

```python
# make 1 column for Name and remove the 2 name columns

df['Name'] = df.insert(0, 'Name', None)
df["Name"] = df["FirstName"] + " " + df["LastName"]
df = df.drop(columns=["FirstName", "LastName"])
df
```

```python
# make Age an int value (why float?)
df["Age"] = df["Age"].astype("Int8")
```

```python
# Flag those with sentences containing Life

# new column for Life sentences (True/False)
df['Life'] = df.insert(5, 'Life', None)
print(list(df))
```

```python

for i,row in df.iterrows(): 
    #print (i, list(row))  # DEBUG: row has values for the row with the index i
    s = row["Sentence"] # sentence
    if 'life' in s.lower():
        df.loc[i, "Life"] = True   
    else:
        df.loc[i, "Life"] = False

df
```

```python
# drop those with Life == "False"
df = df.drop(df[df["Life"] == False].index)
df
```

```python
# Now also drop Life column as it's no longer needed
df = df.drop("Life", axis=1)
```

```python
# add a new column with duration (Exonerated - Convicted)
df['Duration'] = df.insert(5, 'Duration', 0)
df["Duration"] = df["Exonerated"] - df["Convicted"]
df
```

```python
# For DNA use True and False, not DNA and NaN 
for i,row in df.iterrows(): 
    #print (i, list(row))  # DEBUG: row has values for the row with the index i
    if row["DNA"] == "DNA":
        df.loc[i, "DNA"] = True   # short for  blue
    else:
        df.loc[i, "DNA"] = False
df
```

```python
#add median age
age_med = df['Age'].median()

print(age_med) 
```

```python
# Summarize by State:
import pandas as pd
df = pd.read_csv("filtered.csv")

# Convert F- States
for i,row in df.iterrows(): 
    if row["ST"][0:2] == "F-":
            st = row["ST"][2:]  # CH was [3:] !!!!
            df.loc[i, "ST"] =  st 

df["ST"].unique()
```

```python
avgAgebyST = df.groupby("ST")["Age"].mean().round(2)
avgDurbyST = df.groupby("ST")["Duration"].mean().round(2)
DNAprbbyST = df.groupby("ST")["DNA"].mean().round(2)
dfST = pd.concat([avgAgebyST, avgDurbyST, DNAprbbyST], axis=1)
dfST = dfST.rename(columns={'Age':'AvgAge', 'Duration':'AvgDurationYrs', "DNA":"DNAprob"})
dfST
```

```python
numinmatesST = df.value_counts("ST")
numinmatesST = numinmatesST.rename("NumInmates")
dfST = pd.concat([dfST, numinmatesST], axis=1)
dfST
```

```python
race =df.groupby("ST")["Race"].apply(pd.Series.mode)

state_list = []
maj_race_list = []
for st in race.index:
    #print(st, race.loc[st])
    if st[1] == 0:
        state_list.append(st[0])
        maj_race_list.append(race.loc[st])

maj_race_by_state = pd.Series(data=maj_race_list, index=state_list)
maj_race_by_state = maj_race_by_state.rename("MajRace")
dfST = pd.concat([dfST, maj_race_by_state], axis=1)
dfST
```

```python
dfST.index.name = "ST"
dfST.to_csv("StateAggregates.csv")
```

```python
# for each row, make a query for its ST and County into the counties df

for i,row in df.iterrows(): 
        
        # query with ST and County colum 
        query_string = f'state_id == "{row["ST"]}" and county == "{row["County"]}"'  # the " around the values are needed!
        cnty = df_counties.query(query_string)

        if cnty.empty == False:  # we got a hit!
            cnty = cnty.head(1)  # use first of potentially several hits!
            lat, long = float(cnty["lat"]), float(cnty["lng"])
        else: # got no result,  try with county_full?
            query_string = f'state_id == "{row["ST"]}" and county_full == "{row["County"]}"'
            cnty = df_counties.query(query_string)

        if cnty.empty == False:
            cnty = cnty.head(1)  
            lat, long = float(cnty["lat"]), float(cnty["lng"]) 
        else:# Let's see if its  F- country
            if row["ST"][0:2] == "F-":
                st = row["ST"][2:]  # CH was [3:] !!!!
                df.loc[i, "ST"] =  st  # remove F- in data,
                row["ST"] = st  # CH the line above does not overwrite row["ST"] for some reason ...

            # find state and gt it's lat/lon
            state = df_states[df_states["postal_code"] == row["ST"]]
            
            if state.empty == False:  
                lat, long = float(state["lat"]), float(state["lng"]) 
            else:
                print("error for", row["ST"],  row["County"])
                continue
              
        df.loc[i, "Lat"] = lat 
        df.loc[i, "Long"] = long         

df
```
